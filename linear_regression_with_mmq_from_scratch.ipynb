```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self):
        self.N = None
        self.A = None
        self.B = None

    def fit(self, X, y):
        sum_x_train = sum(X)
        sum_y_train = sum(y)

        sum_x_train_square = sum(map(lambda x: x ** 2, X))

        sum_mtply_x_train_y_train = sum([a * b for a, b in zip(X, y)])

        self.N = len(X)
        Bu = self.N * sum_mtply_x_train_y_train - sum_x_train * sum_y_train
        Bd = self.N * sum_x_train_square - (sum_x_train) ** 2
        self.B = Bu / Bd
        self.A = (sum_y_train - self.B * sum_x_train) / self.N

    def predict(self, X):
        return [self._predict(x) for x in X]

    def _predict(self, X):
        return (X * self.B) + self.A # y = bx + a
```

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

fig = plt.figure()
plt.scatter(X[:, 0], y)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
y_predictions = reg.predict(X_test)
```

```python
y_pred_line = reg.predict(X)
```

```python
scatter_train = plt.scatter(X_train, y_train)
scatter_test = plt.scatter(X_test, y_test)
plt.plot(X, y_pred_line, color='red', label='prediction')
plt.show()
```
